```python
import os
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from astropy.visualization import SqrtStretch, ZScaleInterval, MinMaxInterval
from astropy.visualization.mpl_normalize import ImageNormalize
from ccdproc import CCDData

__PATH__ = './data/case_2/'
for __file in os.listdir(__PATH__):
    print(__file)
```

```python
ccd = CCDData.read(os.path.join(__PATH__, 
                                '0359_REFITT-ZTF20ackfpcb-09-11-2020_check.fits'), 
                   unit='adu')
scale = MinMaxInterval()
# scale = ZScaleInterval()

z1, z2 = scale.get_limits(ccd.data)
print(z1, z2)
norm = ImageNormalize(stretch=SqrtStretch())
fig, ax = plt.subplots(figsize=(20,15))
ax.imshow(ccd.data, clim=(z1, z2), cmap=cm.gray, origin='lower', norm=norm,
           interpolation='nearest')
```
