```python
import matplotlib
matplotlib.rcParams['figure.figsize']= [16, 9]
from astropy.modeling import models
import matplotlib.pyplot as plt
import numpy as np

moffat_1d = models.Moffat1D()
print(moffat_1d)
```

```python
x_axis = np.linspace(-10, 10, 100)

for i in np.linspace(-1, 3, 10):
    moffat_1d = models.Moffat1D(amplitude=30, x_0=0, gamma=i, alpha=10)
    print(moffat_1d.fwhm < 0)
    if moffat_1d.fwhm < 0:
        plt.plot(x_axis, moffat_1d(x_axis), color='k')
    else:
        plt.plot(x_axis, moffat_1d(x_axis), label="{}".format(moffat_1d.fwhm), linestyle=':')
    
plt.legend(loc='best')
plt.show()
```
