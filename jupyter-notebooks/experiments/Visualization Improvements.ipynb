```python
import numpy as np

from ccdproc import CCDData
import matplotlib.pyplot as plt

def get_default_limits(ccd, sampling=0.2):
    x, y = ccd.data.shape
    x1 = int(x/2. - sampling * x)
    x2 = int(x/2. + sampling * x)
    y1 = int(y/2. - sampling * y)
    y2 = int(y/2. + sampling * y)
    data_sample = np.mean(ccd.data[:,y1, y2], axis=2)
    plt.imshow(data_sample)
    
#     data_sample = ccd.data[x1:x2, y1:y2]
    z1 = np.mean(data_sample) - 0.5 * np.std(data_sample)
    z2 = np.median(data_sample) +  np.std(data_sample)
    return z1, z2
```

```python
sampling=0.2
x, y = ccd.data.shape
x1 = int(x/2. - sampling * x)
x2 = int(x/2. + sampling * x)
y1 = int(y/2. - sampling * y)
y2 = int(y/2. + sampling * y)

data_sample = np.mean(ccd.data[:,y1, y2], axis=0)

print(data_sample.shape
plt.imshow(data_sample)
```

```python
def get_limits(ccd):
    z1 = np.mean(ccd.data) - 0.5 * np.std(ccd.data)
    z2 = np.median(ccd.data) +  np.std(ccd.data)
    return z1, z2

z1= 3400
z2= 4500

sample_file = 'files/mask_in.fits'

ccd = CCDData.read(sample_file, unit='adu')

dz1, dz2 = get_default_limits(ccd=ccd)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,9))
ax1.imshow(ccd.data, clim=(dz1, dz2), cmap='viridis')
ax2.imshow(ccd.data, clim=(z1, z2), cmap='viridis')
print(dz1,dz2)
```
