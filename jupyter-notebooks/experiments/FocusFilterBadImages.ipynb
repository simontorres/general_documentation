```python
from ccdproc import ImageFileCollection
import numpy as np
import os
from ccdproc import CCDData
folder = '/user/simon/data/soar/focus_fix/2019-11-02'

keywords = ['OBSTYPE',
            'INSTCONF',
            'CAM_TARG',
            'GRT_TARG',
            'FILTER',
            'FILTER2',
            'GRATING',
            'SLIT',
            'WAVMODE',
            'RDNOISE',
            'GAIN',
            'ROI']


ifc = ImageFileCollection(folder, keywords=keywords)

pdifc = ifc.summary.to_pandas()
```

```python
pdifc
```

```python
focus = pdifc[pdifc['OBSTYPE'] == 'FOCUS']
```

```python
focus
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
groups = focus.groupby(['CAM_TARG',
                        'GRT_TARG', 
                        'FILTER', 
                        'FILTER2', 
                        'GRATING', 
                        'SLIT', 
                        'WAVMODE', 
                        'RDNOISE', 
                        'GAIN', 
                        'ROI']).size().reset_index().rename(columns={0: 'count'})

for i in groups.index:
    focus_group = focus[((focus['CAM_TARG'] == groups.iloc[i]['CAM_TARG']) &
                         (focus['GRT_TARG'] == groups.iloc[i]['GRT_TARG']) &
                         (focus['FILTER'] == groups.iloc[i]['FILTER']) &
                         (focus['FILTER2'] == groups.iloc[i]['FILTER2']) &
                         (focus['GRATING'] == groups.iloc[i]['GRATING']) &
                         (focus['SLIT'] == groups.iloc[i]['SLIT']) &
                         (focus['WAVMODE'] == groups.iloc[i]['WAVMODE']) &
                         (focus['RDNOISE'] == groups.iloc[i]['RDNOISE']) &
                         (focus['GAIN'] == groups.iloc[i]['GAIN']) &
                         (focus['ROI'] == groups.iloc[i]['ROI']))]

    _mean = []
    _median = []
    _std = []
    _file_list = focus_group.file.values.tolist()
    for _file in _file_list:
        full_path = os.path.join(folder, _file)
        ccd = CCDData.read(full_path, unit='adu')
        spectral_profile = np.mean(ccd.data, axis=0)
        z1 = np.mean(ccd.data) - 0.5 * np.std(ccd.data)
        z2 = np.median(ccd.data) + np.std(ccd.data)
        fig, ax =  plt.subplots(figsize=(16, 2))
        ax.plot(spectral_profile)
#         im = ax.imshow(ccd.data, clim=(z1,z2), cmap='viridis')
#         divider = make_axes_locatable(ax)
#         cax = divider.append_axes('top', size='10%', pad=0.3)
#         plt.colorbar(im, cax=cax, orientation='horizontal')
        fig.suptitle(_file)
        plt.show()
        _mean_ = np.mean(ccd.data)
        _median_ = np.median(ccd.data)
        _mean.append(_mean_)
        _median.append(_median_)
        _std_ = np.std(ccd.data)
        _std.append(_std_)
        print("Mean: {} Median: {} Stdev: {}".format(_mean_, _median_, _std_))
#         print(np.mean(ccd.data), np.median(ccd.data), np.std(ccd.data))
    fig, ((ax1, ax2 ), (ax3, ax4)) = plt.subplots(2,2, figsize=(16, 9))
    fig.suptitle("{} {} {}".format(groups.iloc[i]['WAVMODE'],groups.iloc[i]['FILTER'], groups.iloc[i]['FILTER2']))
    ax1.plot(_mean)
    ax1.set_title('Mean')
    ax2.plot(_median)
    ax2.set_title('Median')
    ax3.plot(_std)
    ax3.set_title('STD')
    ax4.hist(_std)
    plt.show()

```
