```python
import numpy as np
import matplotlib.pyplot as plt
from ccdproc import CCDData
```

```python
# image = '/user/simon/data/soar/work/paula/2017-07-19/RED/cfzsto_0134_night01_stripe23.fits'
image = '/user/simon/data/soar/work/paula/2017-07-19/RED/cfzsto_0073_night01.fits'
ccd = CCDData.read(image)

plt.imshow(ccd.data, clim=(10, 70))
plt.title(image.split('/')[-1])
plt.xlabel('Dispersion Axis')
plt.ylabel('Spatial Axis')
plt.show()
```
