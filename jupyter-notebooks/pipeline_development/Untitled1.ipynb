```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [9,6]
```

```python
def f(t):
    return (t - 2) ** 2

t1 = np.arange(0.0, 5.0, 0.5)
t2 = np.arange(0.0, 5.0, 0.02)

plt.rcParams['font.size'] = 24
fig, ax = plt.subplots(num='poinnts')
ax.set_title('A simple plot')

ax.plot(t1, f(t1), 'ro', label='Points')
ax.plot(t2, f(t2), 'k', label='Line')
ax.set_ylabel("$f(t) = (x -2) ^ 2$")
ax.set_xlabel("t [s]")
ax.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.rcParamsDefault
plt.rcParams['figure.figsize'] = [10,7]
x = np.arange(0.0, 5.0, 0.5)
y = (x - 2) ** 2
plt.style.use('seaborn')

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(223)
ax3 = fig.add_subplot(224)

ax1.plot(x, y, 'C0o')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-1, 5)
ax1.set_title("Axis 1 (211)")

ax2.plot(x, y, "C1s")
ax2.set_title("Axis 2 (223)")

ax3.plot(x, y, "C2:")
ax3.set_title("Axis 3 (224)")

fig.tight_layout()
plt.show()
```
