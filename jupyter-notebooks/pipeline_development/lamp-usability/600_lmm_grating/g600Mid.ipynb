Ar     | No     | Not enough lines                      | Discarded
HgAr   | Yes    | Prefer HgArNe if you want more lines  | Yes
CuHeAr | No     | Not enough lines                      | Discarded
HgArNe | Yes    | Best, though coverage is not ideal    | Yes

```python
from ccdproc import CCDData

import glob
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (14, 6)

def _plot_lamps(file_list):
    for _file in file_list:
        ccd = CCDData.read(_file, unit='adu')
        fig, ax = plt.subplots(num=_file)
        ax.set_title(_file)
        ax.plot(ccd.data, label='Lamp: {:s} exposed {:.1f} Seconds'.format(ccd.header['OBJECT'], ccd.header['EXPTIME']))
        ax.legend()
        plt.show()
```

```python
files_list = glob.glob('files/*600Mid*fits')
_plot_lamps(file_list=files_list)
```
