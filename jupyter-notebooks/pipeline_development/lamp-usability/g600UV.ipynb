# Analysis of Comparison Lamps Usability

# 600 l/mm UV
Lamp   | Usable | Comments
---    | ---    | ---
Ne     | No     | Few lines and high background
FeHeAr | No     | Bad line distribution and signal
Ar     | No     | Bad line distribution and signal
HgAr   | Yes    | Best
CuHeAr | Yes    | Not recommended
HgArNe | Yes    | Prefer HgAr

```python
from ccdproc import CCDData

import glob
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (14, 6)

def _plot_lamps(file_list):
    for _file in file_list:
        ccd = CCDData.read(_file, unit='adu')
        fig, ax = plt.subplots(num=_file)
        ax.set_title(_file)
        ax.plot(ccd.data, label='Lamp: {:s} exposed {:.1f} Seconds'.format(ccd.header['OBJECT'], ccd.header['EXPTIME']))
        ax.legend()
        plt.show()
    
```


## 600UV

```python
files_list = glob.glob('files/*600UV*fits')
_plot_lamps(file_list=files_list)
```
