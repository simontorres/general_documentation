```python
import matplotlib.pyplot as plt
from astropy.modeling import models
import numpy as np


gaussian = models.Gaussian1D(amplitude=10, mean=5, stddev=2)
moffat = models.Moffat1D(amplitude=10, x_0=5, gamma=2)

fig, ax = plt.subplots(figsize=(18,10))
x_axis = np.linspace(0, 10, 100)
plt.plot(x_axis, gaussian(x_axis), label='Gaussian')
plt.plot(x_axis, moffat(x_axis), label='Moffat')
plt.legend()
plt.show()
```

```python
moffat = models.Moffat1D(amplitude=10, x_0=5, gamma=2)
import numpy as np
alphas = np.linspace(0, 10, 100)
fig, ax = plt.subplots(figsize=(18,10))

ax.set_title('Alpha variable')
x_axis = np.linspace(0, 10, 100)

for y in alphas:
    moffat.alpha.value = y
    ax.plot(x_axis, moffat(x_axis), label="{:.1f}".format(y))
plt.legend()
plt.show()
    
```

```python
moffat = models.Moffat1D(amplitude=10, x_0=5, alpha=10)
import numpy as np
gammas = np.linspace(0, 10, 100)
fig, ax = plt.subplots(figsize=(18,10))

ax.set_title('Gamma variable')
x_axis = np.linspace(0, 10, 100)

for y in gammas:
    moffat.gamma.value = y
    ax.plot(x_axis, moffat(x_axis), label="{:.1f}".format(y))
plt.legend()
plt.show()
    
```
