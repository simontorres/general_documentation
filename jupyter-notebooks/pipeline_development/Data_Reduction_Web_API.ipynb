# Reducing data with a web API implementation

Initially, there are some steps required to reduce the data. Every step has to be an API endpoint.
This API is private, the user interface will be the web app that has an authenticated public API.

This is highly experimental so if something does not work it will change.

# Object Model

- Sessions
- file
- Image
- Spectra




## API endpoints summary

(do not use verbs)

### Sessions

`POST /api/session`

`POST /api/session/overscan`

`POST /api/session/trimsec`

`POST /api/session/slittrim`

### Create Calibrations

`POST /api/calibrations/bias`

`POST /api/calibrations/flats`

### Retrieve data

`GET /api/image/overscan`

`GET /api/image/trimsec`

`GET /api/image/slittrim`

### Image Reduction

`POST /api/image/overscan`

`POST /api/image/trim`

`POST /api/image/bias`

`POST /api/image/flat`

`POST /api/image/creject`

### Spectral Reduction

`POST /api/spectra/identify`

`POST /api/spectra/trace`

`POST /api/spectra/extract`

### Spectral Calibration

`POST /api/spectra/wavelength`

`POST /api/spectra/flux`


# Initialization

There has to be an initialization process, when exactly it happens it has to be defined, but in principle could happen at the very begining or after all the calibration data has been taken. The should be initialized which will return a session ID which has to be used in every call afterwards

`/api/session`

the data should be parsed is the directory where data is or 

```json
{'api_version': '1.0.0',
 'observing_date': '20190910',
 'raw_data_path': '/home/user/data',
 'reduced_data_path': '/home/user/reduced_data',
 'advanced_settings': 'none'}

```

returns

```json
{'api_version': '1.0.0',
 'session_id': 'someid' }
```

# Calibrations: Master Bias

It should work be able to automatically discover all images or receive a list of files.

`POST /api/calibrations/bias`


```json
{'auto_discover': 'true',
 'file_list': []}

```

`auto_discover` and `file_list` should be mutually exclusive with one of the having priority in case both are provided.

# Calibrations: Master Flats

Same as Master bias

`POST /api/calibrations/flats`

```json
{'auto_discover': 'true',
 'file_list': []}

```

# Reduction

## Overscan

`POST /api/reduce/overscan`

## Trim image

`POST /api/reduce/trim`

```json
{'trim_section':'[1:2,1:100]',
 'trim_type': 'slit'}
```

## Bias Correction

`POST /api/reduce/bias`

```json
{'master_bias':'master_bias.fits'}
```

## Flat Correction

`POST /api/reduce/flat`

```json
{'master_flat':'master_flat.fits'}
```

## Cosmic Ray Rejection

`POST /api/reduce/creject`

```json
{'method': 'lacosmic'}
```
